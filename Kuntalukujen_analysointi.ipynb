```python
#Luodaan listat halutuille sarakkeille

kunnat = []
asukkaat = []
kasvu = []
työllisyys = [] 

file = open("Kuntien avainluvut.csv","r")
skip = next(file)
print(skip)
for row in file:
    row = row.strip()
    parts = row.split(";")
    print(parts)
    kunnat.append(parts[0])
    asukkaat.append(int(parts[1]))
    kasvu.append(float(parts[2]))
    työllisyys.append(float(parts[3]))
file.close()

print(kunnat)
print(asukkaat)
print(kasvu)
print(työllisyys)
```

```python
#Mikä on pääkaupunkiseudun asiakasluku?
print(kunnat.index("Helsinki"))
luku1 = int(asukkaat[34])
print(luku1)
print(kunnat.index("Vantaa"))
luku2 = int(asukkaat[292])
print(luku2)
print(kunnat.index("Espoo"))
luku3 = int(asukkaat[12])
print(luku3)
print(kunnat.index("Kauniainen"))
luku4 = int(asukkaat[84])
print(luku4)
sum = luku1 + luku2 + luku3 + luku4
print("Pääkaupunkiseudun asiakasluku on:", sum)
```

```python
#Lasketaan numpyn avulla samat asiat
import numpy as np
import csv

asukkaat = []

with open("Kuntien avainluvut.csv") as file:
    csv_file = csv.reader(file,delimiter=";")
    otsikko = next(csv_file)
    for rivi in csv_file:
        asukkaat.append(int(rivi[1]))
        
asukkaat_np = np.array(asukkaat)

print("Suurin asukasmäärä on:", np.max(asukkaat_np))
print("Pienin asukasmäärä on:", np.min(asukkaat_np))
print("asukkaiden keskimäärä on:", np.mean(asukkaat_np))
print("Asukasmäärän mediaani on:", np.median(asukkaat_np))
```

```python
#Visualisoidaan osa datasta matplotlibillä

%matplotlib inline
import matplotlib.pyplot as plt

asukkaat = []
työllisyysaste = []
huoltosuhde = []

with open("Kuntien avainluvut.csv") as file:
    csv_file = csv.reader(file,delimiter=";")
    otsikko = next(csv_file)
    for rivi in csv_file:
        asukkaat.append(float(rivi[1]))
        työllisyysaste.append(float(rivi[3]))
        huoltosuhde.append(float(rivi[4]))
        
plt.figure() 
plt.plot(asukkaat)
plt.plot(työllisyysaste)
plt.show()
```
